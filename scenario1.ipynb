```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors
import pandas as pd

# Matrice binaire : 1 = NA, 0 = pas NA
na_matrix = df.isna().astype(int).values

# Palette binaire : blanc = 0, rouge = 1
cmap = mcolors.ListedColormap(["white", "red"])

plt.figure(figsize=(14, 10))

# Heatmap
plt.imshow(na_matrix, cmap=cmap, aspect="auto")

# --- Axe X : noms des colonnes ---
plt.xticks(
    ticks=np.arange(len(data.columns)),
    labels=data.columns,
    rotation=90
)

# --- Axe Y : noms des pays ---
plt.yticks(
    ticks=np.arange(len(data)),
    labels=data["Country"],
    fontsize=6
)

# --- Quadrillage ---
plt.grid(color='black', linestyle='-', linewidth=0.3)

# Ajuster le quadrillage pour qu’il s’aligne sur les cellules
plt.gca().set_xticks(np.arange(-0.5, len(data.columns), 1), minor=True)
plt.gca().set_yticks(np.arange(-0.5, len(data), 1), minor=True)
plt.grid(which='minor', color='black', linestyle='-', linewidth=0.3)
plt.grid(False)  # enlève la grille principale, garde la fine

plt.colorbar(label="1 = NA (rouge), 0 = pas NA (blanc)")
plt.title("Heatmap missing data (NA)")

plt.tight_layout()
plt.show()
```

```python
# Compter les NA par ligne
na_par_pays = df.isna().sum(axis=1)

# Ajouter le nom du pays
resultat = df[["Country"]].copy()
resultat["nb_NA"] = na_par_pays

# Trouver le pays avec le maximum de NA
pays_max_na = resultat.loc[resultat["nb_NA"].idxmax()]

print(pays_max_na)
```
